Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python

# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
df_converted_comp = pd.read_sql_query("SELECT ConvertedComp FROM master", conn)

import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df_converted_comp['ConvertedComp'].dropna(), bins=30, edgecolor='black')
plt.title('Distribution of ConvertedComp (Annual USD Salaries)')
plt.xlabel('Annual Compensation (USD)')
plt.ylabel('Number of Respondents')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df_age = pd.read_sql_query("SELECT Age FROM master", conn)
plt.figure(figsize=(10, 6))
sns.boxplot(y=df_age['Age'].dropna())
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df_age_workhours = pd.read_sql_query("SELECT Age, WorkWeekHrs FROM master", conn)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_age_workhours['Age'], y=df_age_workhours['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
df_data = pd.read_sql_query("SELECT WorkWeekHrs, CodeRevHrs, Age FROM master", conn)
df_data = df_data.sort_values(by='Age')

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_data, x='WorkWeekHrs', y='CodeRevHrs', size='Age', 
                sizes=(20, 500),  # Setting the range for bubble sizes
                alpha=0.5, hue='Age', palette='viridis')

plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Age as bubble size)')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.legend(title='Age')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
df_desired_databases = pd.read_sql_query("SELECT * FROM DatabaseDesireNextYear", conn)
database_desire_counts = df_desired_databases['DatabaseDesireNextYear'].value_counts().sort_values(ascending=False)

top_5_databases = database_desire_counts.head(5)

# Plotting the pie chart for the top 5 desired databases
plt.figure(figsize=(10, 7))
top_5_databases.plot.pie(autopct='%1.1f%%', startangle=90, labels=top_5_databases.index)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.ylabel('')  # Removing the default ylabel
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df_age_group = pd.read_sql_query("SELECT WorkWeekHrs, CodeRevHrs, Age FROM master WHERE Age BETWEEN 30 AND 35", conn)
median_values = df_age_group.groupby('Age').median()

ax = median_values.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.xticks(rotation=0) 
plt.legend(title='Activity')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_age_range = pd.read_sql_query("SELECT ConvertedComp, Age FROM master WHERE Age BETWEEN 45 AND 60", conn)
median_comp_values = df_age_range.groupby('Age').median()

plt.figure(figsize=(12, 7))
median_comp_values['ConvertedComp'].plot(kind='line', marker='o')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(list(range(45, 61)))
plt.tight_layout()
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
df_mainbranch = pd.read_sql_query("SELECT MainBranch FROM master", conn)

# Counting the frequency of each category in the 'MainBranch' column
mainbranch_counts = df_mainbranch['MainBranch'].value_counts()

plt.figure(figsize=(8, 5))
bars = plt.barh(mainbranch_counts.index, mainbranch_counts.values, color='skyblue')

plt.xlabel('Number of Respondents')
plt.title('Distribution of MainBranch of Respondents')

# Placing the bar label aligned slightly away from the y-axis
for bar, label in zip(bars, mainbranch_counts.index):
    plt.text(100,  # Adjusted X position for a small margin
             bar.get_y() + bar.get_height()/2,            # Y position to center on the bar
             label,                                       # Text value
             ha='left', va='center',
             color='black', fontsize=10)

plt.gca().invert_yaxis()  # Displaying the category with the highest count at the top
plt.gca().axes.get_yaxis().set_visible(False)  # Hiding the default y-axis labels
plt.tight_layout()
plt.show()
```
